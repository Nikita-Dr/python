```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi, cos, sin, radians, degrees, e as exp, sqrt
x=[]
y=[]
z=[]
R=0.05
alpha=pi/2
betta=pi/3
q=5
m=0.05
v=300
B=100
dt=0.000001
v0x=v*sin(alpha)*cos(betta)
v0y=v*sin(alpha)*sin(betta)

x.append(0.0)
y.append(R)
z.append(0.0)

vz=v*cos(alpha)
vx=[]
vy=[]
vx.append(v0x-(q*B*v0y/m)*dt/2)
vy.append(v0y-(-q*B*v0x/m)*dt/2)

for i in range (1,2000):
  vx.append(vx[i-1]-(q*B*vy[i-1]/m)*dt)
  vy.append(vy[i-1]-(-q*B*vx[i]/m)*dt)
  x.append(x[i-1]+vx[i]*dt)
  y.append(y[i-1]+vy[i]*dt)
  z.append(z[i-1]+vz*dt)
fig1 = plt.figure(figsize=(6, 6))
plt.xlabel("x")
plt.ylabel("y")
graph1 = plt.plot(x,y)
grid1 = plt.grid(True)
fig2 = plt.figure(figsize=(6, 6))
plt.xlabel("x")
plt.ylabel("z")
graph2 = plt.plot(x,z)
grid2 = plt.grid(True)
fig3 = plt.figure(figsize=(6, 6))
plt.xlabel("y")
plt.ylabel("z")
graph3 = plt.plot(y,z)
grid3 = plt.grid(True)
fig4 = plt.figure(figsize=(10, 10))
ax = fig4.add_subplot(projection='3d')
ax.plot(x, y, z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```
